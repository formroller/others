```python
# import os 
# os.chdir('./medical')

import pandas as pd
import numpy as np

from collections import Counter
```

```python
# 데이터 load
df = pd.read_csv('portal3.csv', index_col=False, encoding='cp949',usecols=['fname','provide'])
```

```python
## setting
rows = np.arange(len(df))
df['title'] = [df['fname'].str.split(' ')[i][0] for i in range(len(df)) ]  # 제목 -> 지역, 기관 분기

second = []
for i in df['fname']:
    if len(i.split(' ')) >= 2:
        second.append(i.split(' ')[1])
    else:
        second.append('')
```

```python
Counter(second)
```

```python
# =============================================================================
# 지역(local) 설정 loc
# =============================================================================
loc=[]
for i in df['title']:
    if i.endswith('시'):
        loc.append(i)
    elif i.endswith('도'):
        loc.append(i)
    elif i.endswith('구'):
        loc.append(i)
    elif i.endswith('군'):
        loc.append(i)
    else:
        loc.append('')
df['local'] = loc
```

```python
# sub_loc
sub_loc=[]
for i in second:
    if i.endswith('시'):
        sub_loc.append(i)
    elif i.endswith('군'):
        sub_loc.append(i)
    elif i.endswith('구'):
        sub_loc.append(i)
    else:
        sub_loc.append('')
df['sub_loc'] = sub_loc
```
